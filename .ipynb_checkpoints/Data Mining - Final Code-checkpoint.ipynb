# Discretization

```python
def equal_width_binning(column):
    # Define the custom bin edges
    bin_edges = [0, 0.33, 0.67, 1]
    
    # Define the corresponding bin labels
    bin_labels = [1, 2, 3]
    
    return pd.cut(column, bins=bin_edges, labels=bin_labels, include_lowest=True)
```

```python
columns_to_bin = ['BALANCE_FREQUENCY', 'PURCHASES_FREQUENCY','ONEOFF_PURCHASES_FREQUENCY','PURCHASES_INSTALLMENTS_FREQUENCY','CASH_ADVANCE_FREQUENCY' ] 

# Apply equal width binning function to each column
for col in columns_to_bin:
    df[col] = equal_width_binning(df[col])
```

```python
def binarize(threshold,column_name):
    
    df[column_name] = df[column_name].apply(lambda x: 1 if x > threshold else 0)
    df[column_name] = df[column_name].astype('category')
    
    return df
```

# K Means Clustering

```python
from sklearn.cluster import KMeans
max_clusters = 10
```

```python
def find_optimal_clusters_using_elbow(X, max_k):
    wcss = []  # List to store the within-cluster sum of squares (WCSS) for different values of k

    # Fit k-means clustering for different values of k
    for k in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)  # Append the WCSS for the current value of k to the list
    
    # Plot the elbow curve
    plt.plot(range(1, max_k + 1), wcss, marker='o')
    plt.title('Elbow Method')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.show()
    
find_optimal_clusters_using_elbow(X_pca, max_clusters)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Step 1: Find optimal number of clusters using silhouette score
silhouette_scores = []

for n_clusters in range(2, max_clusters+1): 
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X_pca)
    silhouette_avg = silhouette_score(X_pca, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot silhouette score vs number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs Number of Clusters')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
```

```python
optimal_n_clusters = np.argmax(silhouette_scores) + 2  # Add 2 because range starts from 2
print(optimal_n_clusters)
```

```python
optimal_n_clusters = 4
# Step 2: Find optimal initial centroids
kmeans = KMeans(n_clusters=optimal_n_clusters, init='k-means++', random_state=42)
kmeans.fit(X_pca)
optimal_initial_centroids = kmeans.cluster_centers_
```

```python
# Step 3: Perform actual clustering with optimal parameters
start_time = time.time()
kmeans = KMeans(n_clusters=optimal_n_clusters, init=optimal_initial_centroids, n_init=1, random_state=42)
cluster_labels = kmeans.fit_predict(X_pca)
end_time=time.time()
```

```python
from scipy.spatial.distance import pdist, squareform

def total_sum_of_squares(X):
    pairwise_distances = squareform(pdist(X, 'sqeuclidean'))
    tss = np.sum(pairwise_distances)
    return tss


TSS = total_sum_of_squares(X_pca)
print("Total Sum of Squares (TSS):", TSS)

def within_cluster_sum_of_squares(X, cluster_labels, centroids):
    squared_distances = np.sum((X - centroids[cluster_labels])**2, axis=1)
    wcss = np.sum(squared_distances)
    return wcss

WCSS = within_cluster_sum_of_squares(X_pca, cluster_labels, kmeans.cluster_centers_)
print("Within-Cluster Sum of Squares (WCSS):", WCSS)

def percentage_variance_explained(TSS, WCSS):
    return (TSS - WCSS) / TSS * 100

PVE = percentage_variance_explained(TSS, WCSS)
print("Percentage of Variance Explained:", PVE)
```

```python
from sklearn.neighbors import NearestNeighbors
def get_kdist_plot(X=None, k=None, radius_nbrs=1.0):

    nbrs = NearestNeighbors(n_neighbors=k, radius=radius_nbrs).fit(X)

    # For each point, compute distances to its k-nearest neighbors
    distances, indices = nbrs.kneighbors(X) 
                                       
    distances = np.sort(distances, axis=0)
    distances = distances[:, k-1]

    # Plot the sorted K-nearest neighbor distance for each point in the dataset
    plt.figure(figsize=(8,8))
    plt.plot(distances)
    plt.xlabel('Points/Objects in the dataset', fontsize=12)
    plt.ylabel('Sorted {}-nearest neighbor distance'.format(k), fontsize=12)
    plt.grid(True, linestyle="--", color='black', alpha=0.4)
    plt.show()
    plt.close()


k = 2 * X_pca.shape[-1] - 1 # k=2*{dim(dataset)} - 1
print(k)
get_kdist_plot(X=X_pca, k=k)
```

# Hyperparameter Tuning

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Define range of values for eps and min_samples
eps_range = np.linspace(1, 10, num=34, endpoint=True) 
min_samples_range = range(5, 20)
desired_number_of_clusters = 4

# Initialize variables to track best parameters and silhouette score
best_eps = None
best_min_samples = None
best_silhouette_score = -1  # Initialize with a low value

# Iterate over eps and min_samples values
for eps in eps_range:
    for min_samples in min_samples_range:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        cluster_labels = dbscan.fit_predict(X_pca)
        
        silhouette = silhouette_score(X_pca, cluster_labels)
        
        unique_clusters = len(np.unique(cluster_labels[cluster_labels != -1]))
        if unique_clusters < desired_number_of_clusters:
            continue
        # Update best parameters if current silhouette score is higher
        if silhouette > best_silhouette_score and unique_clusters==desired_number_of_clusters:
            best_silhouette_score = silhouette
            best_eps = eps
            best_min_samples = min_samples

# Print the best parameters and corresponding silhouette score
print(f"Optimal eps: {best_eps}")
print(f"Optimal min_samples: {best_min_samples}")
print(f"Max Silhouette Score: {best_silhouette_score}")
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting

# Define the DBSCAN parameters
eps = best_eps  # The maximum distance between two samples to be considered in the same neighborhood
min_samples = best_min_samples  # The number of samples in a neighborhood for a point to be considered a core point

dbscan = DBSCAN(eps=eps, min_samples=min_samples)
cluster_labels = dbscan.fit_predict(X_pca)

cluster_labels_no_noise = cluster_labels[cluster_labels != -1]

unique_labels = set(cluster_labels)
num_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)  # Number of clusters
num_outliers = list(cluster_labels).count(-1)  # Number of outliers (noise points)

# Visualize the clusters and outliers in a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for each cluster label (excluding noise)
for label in unique_labels:
    if label == -1:
        # Plot outliers (noise) in black color
        indices = cluster_labels == label
        ax.scatter(X_pca[indices, 0], X_pca[indices, 1], X_pca[indices, 2], c='k', marker='o', label='Outliers')
    else:
        # Plot points belonging to each cluster with different colors
        indices = cluster_labels == label
        ax.scatter(X_pca[indices, 0], X_pca[indices, 1], X_pca[indices, 2], label=f'Cluster {label}')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('DBSCAN Clustering of 3D Data')

ax.legend()

# Print number of clusters and outliers
print(f"Number of Clusters: {num_clusters}")
print(f"Number of Outliers (Noise Points): {num_outliers}")

# Show the plot
ax.view_init(elev=20, azim=60)
plt.show()
```
